```python
miss = {'Meowstic Male': 'https://www.smogon.com/dex/sv/pokemon/meowstic-m/', 'Meowstic Female': 'https://www.smogon.com/dex/sv/pokemon/meowstic-f/', "Sirfetch'd": 'https://www.smogon.com/dex/ss/pokemon/sirfetchd/', 'Mr. Rime': 'https://www.smogon.com/dex/ss/pokemon/mr-rime/' }
```

```python
for i, (key, value) in enumerate(miss.items()):
    target_url = value
    
    print(i, key, value)
    
    options= Options()
    options.add_argument("-headless")
    driver = webdriver.Firefox(options=options)
    driver.execute_script(f"location.href='{target_url}';")
    time.sleep(5)
    resp = driver.page_source
    soup = BeautifulSoup(resp)

    driver.close()

    container = soup.find("div", {"class": "DexTable is-even"})

    moves_list = []
    for move in container:
        moves_dict = {}
        move_table = move.contents

        for i in range(7):
            if i == 0:
                moves_dict[move_table[i].contents[0].get_text()] = {}
                temp_key = move_table[i].contents[0].get_text()
                temp = moves_dict
            elif i == 1:
                temp[temp_key]["Type"] = move_table[i].contents[0].get_text()
            elif i == 2:
                temp[temp_key]["Damage"] = move_table[i].contents[0]['class'][1]
            elif i == 3:
                temp[temp_key]["Power"] = move_table[i].contents[1].get_text()
            elif i == 4:
                temp[temp_key]["Accuracy"] = move_table[i].contents[1].get_text()
            elif i == 5:
                temp[temp_key]["PP"] = move_table[i].contents[1].get_text()
            elif i == 6:
                temp[temp_key]["Description"] = move_table[i].contents[0].get_text()  

    moves_list.append(moves_dict)
        
    final_df.loc[final_df['name'] == key, "moves"] = moves_list
```
